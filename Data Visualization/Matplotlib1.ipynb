# Histogram

```python
ml_stu = np.random.randint(18,45,(100))
py_stu = np.random.randint(15,100,(100))
```

```python
plt.figure(figsize=(16,9))
bins = [15,20,25,30,35,40,45]
plt.hist([ml_stu,py_stu],bins,rwidth=0.8,histtype='bar',color=['m','b'],label=['ML Students','PY Students'])
# plt.hist(py_stu,bins,rwidth=0.8,histtype='bar',orientation='vertical',color='b',label='PY Students')
plt.title("PY and ML Students",fontsize=20)
plt.xlabel('Age',fontsize=20)
plt.ylabel('Level',fontsize=20)
plt.legend()
plt.show()
```

# Bar

```python
classes = ['Python' , 'Machine Learning' , 'Artificial Intelligence', 'Flask' , 'Java' , 'HTML/CSS']
class1 = [10,20,40,60,80,90]
class2 = [40,50,80,20,30,90]
class3 = [25,65,75,95,15,35]
```

```python
plt.figure(figsize=(16,9))
index = np.arange(len(classes))
width=0.2

plt.bar(index,class1,width,color='b')
plt.bar(index+width,class2,width,color='r')
plt.bar(index+width+width,class3,width,color='y')

plt.xticks(index+width,classes,rotation=33)
plt.show()
```

# Pie Chart

```python
classes = ['Python' , 'Machine Learning' , 'Artificial Intelligence', 'Flask' , 'Java' , 'HTML/CSS']
class1 = [25,65,75,95,45,35]
```

```python
plt.figure(figsize=(16,9))

explode = [0,0.1,0.2,0,0,0]
colors = ['c','b','g','r','m','y'] 
textprops = {'fontsize':15}
plt.pie(class1,labels=classes,explode=explode,colors=colors,autopct='%0.2f%%',shadow=True,textprops=textprops )
plt.show()
```

```python
coloring = ['r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w']
labels = np.ones(20)
```

```python
plt.figure(figsize=(16,9))
plt.pie([1],colors='b',radius=2.02)
plt.pie(labels,colors=coloring,radius=2.0)
plt.pie([1],colors='g',radius=1.8)
plt.pie([1],colors='y',radius=1.6)
plt.pie([1],colors='m',radius=1.4)
plt.pie([1],colors='b',radius=1.2)
plt.pie([1],colors='c',radius=1.0)
plt.pie(labels,colors=coloring,radius=0.8)
plt.pie([1],colors='k',radius=0.6)
plt.pie([1],colors='g',radius=0.4)
plt.pie(labels,colors=coloring,radius=0.2)
plt.savefig('Pie_chart')
plt.show()
```
