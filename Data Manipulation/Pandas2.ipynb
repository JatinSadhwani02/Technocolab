# Merge

```python
d1 = pd.DataFrame({'ID':[1,2,3,4,5],'Colors':['Red','Blue','Green','Yellow','Pink']})
d2 = pd.DataFrame({'ID':[1,2,3,4,5],'Fruits':['Orange','Banana','Grapes','Mango','Apple']})
```

```python
d1
```

```python
d2
```

```python
pd.merge(d1,d2,on='ID')
```

```python
d3 = pd.DataFrame({'ID':[1,2,3,4,5],'Colors':['Red','Blue','Green','Yellow','Pink']})
d4 = pd.DataFrame({'ID':[1,2,3,6,7],'Fruits':['Orange','Banana','Grapes','Mango','Apple']})
```

```python
pd.merge(d3,d4,on="ID")
```

```python
pd.merge(d3,d4,left_index=True,right_index=True)
```

```python
pd.merge(d3,d4,left_index=True,right_index=True,suffixes=['_Colors_Index','_Fruits_Index'])
```
