## From the given ‘Iris’ dataset, predict the optimum number of clusters and represent it visually.

## import required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

## load data

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
iris.shape
```

## drop unneccesory features

```python
iris.drop(columns=['Id','Species'],axis=1,inplace=True)
```

```python
iris
```

## Find optimal number of clusters using Elbow Method

```python
x = iris.iloc[:,[0,1,2,3]].values
wcss_list = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i , init='k-means++',random_state=51)
    kmeans.fit(x)
    wcss_list.append(kmeans.inertia_)
    
plt.plot(range(1,11),wcss_list)
plt.title('The Elbow Method Graph')
plt.xlabel('The number of clusters')
plt.ylabel('WCSS List')
plt.show()
```

## We can see above graph that elbow point is at 3. So the number of clusters here will be 3

```python
kmeans = KMeans(n_clusters=3, init='k-means++',random_state=51)
kmeans.fit(x)
pred = kmeans.predict(x)
```

```python
pred
```

## Add a cluster feature in iris dataset to show which cluster does the particular feature belong to

```python
iris['cluster'] = pred
iris
```

```python
iris['cluster'].unique()
```

```python
x[pred==0,1]
```

```python
kmeans.cluster_centers_[:,0]
```

## Plotting a scatter plot showing the cluster

```python
plt.figure(figsize=(10,10))
plt.title('K Means Clustering')
plt.scatter(x[pred==0,0],x[pred==0,1],c='red',label='Iris-setosa')
plt.scatter(x[pred==1,0],x[pred==1,1],c='aqua',label='Iris-virginica')
plt.scatter(x[pred==2,0],x[pred==2,1],c='black',label='Iris-versicolor')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1],s=100, c = 'yellow', label = 'Centroids')
plt.legend(loc='best')
```
